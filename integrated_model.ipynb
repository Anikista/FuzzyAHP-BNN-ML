```python
import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(data_1):
    outliers=[]
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```

```python
value = set(raw_data['Vendor'])
for i in value:
    print(i)
    raw_dataa=raw_data[raw_data['Vendor'] == i]
    dataset=raw_dataa['Length(cm)']
    outlier_datapoints = detect_outlier(dataset)
    for j in outlier_datapoints:
        filtered_data = raw_data[(raw_data['Vendor'] == i) & (raw_data['Length(cm)'] == j)]
        raw_data.drop(filtered_data.index, inplace=True)
```
